```python
# Creating EGA
# per-pack difference (actual - target)
df['ega_per_pack'] = df['pack_grammage'] - df['target_weight']


# total giveaway per record (per-pack * count)df
df['ega_total_gm'] = df['ega_per_pack'] * df['actual_count']


# percent giveaway relative to target (per pack)
df['ega_total_per'] = (df['ega_per_pack'] / df['target_weight']) * 100

df
```

```python
import matplotlib.pyplot as plt

target = df['target_weight']
actual = df['actual_weight']   # or df['pack_grammage'] if thatâ€™s per-pack

plt.figure(figsize=(12,5))

# Color code: green = overfill, red = underfill
colors = ['green' if w >= target.iloc[0] else 'red' for w in actual]

plt.bar(range(len(actual)), actual, color=colors, alpha=0.7, label='Actual Pack Weight')

# Target line
plt.axhline(y=target.iloc[0], color='blue', linestyle='--', linewidth=2, label='Target Weight')

plt.title('Actual Pack Weights vs Target')
plt.xlabel('Pack Index')
plt.ylabel('Weight (gm)')
plt.legend()
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

target = df['target_weight'].iloc[0]   # target is constant
actual = df['actual_weight']

# Group into chunks of 500 and take mean
chunk_size = 500
avg_weights = actual.groupby(np.arange(len(actual)) // chunk_size).mean()

plt.figure(figsize=(10,5))

# Color: green if avg >= target, red if avg < target
colors = ['green' if w >= target else 'red' for w in avg_weights]

plt.bar(range(len(avg_weights)), avg_weights, color=colors, alpha=0.7)

# Target line
plt.axhline(y=target, color='blue', linestyle='--', linewidth=2, label='Target Weight')

plt.title(f'Average Pack Weights (per {chunk_size} packs) vs Target')
plt.xlabel(f'Batch of {chunk_size} Packs')
plt.ylabel('Average Weight (gm)')
plt.legend()
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

```python
# over fill
positive_ega = df[df['ega_total_gm'] > 0]
positive_count = positive_ega.shape[0]

# underfill
negative_ega = df[df['ega_total_gm'] < 0]
negative_count = negative_ega.shape[0]

# Zero EGA (exact match=actual)
zero_ega = df[df['ega_total_gm'] == 0]
zero_count = zero_ega.shape[0]

# Print results
print(f"Number of overfilled records (ega_total_gm > 0): {positive_count}")
print(f"Number of underfilled records (ega_total_gm < 0): {negative_count}")
print(f"Number of exact match records (ega_total_gm = 0): {zero_count}")
```

```python
categories = ['Overfill', 'Underfill', 'Exact Match']
counts = [positive_count, negative_count, zero_count]

plt.figure(figsize=(8,5))
plt.bar(categories, counts, color=['green', 'red', 'blue'])
plt.title('EGA Total Giveaway Distribution', fontsize=14)
plt.xlabel('Category', fontsize=12)
plt.ylabel('Number of Records', fontsize=12)
plt.show()
```
